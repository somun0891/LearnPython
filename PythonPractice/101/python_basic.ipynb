```python
from datetime import datetime,date,time,timedelta

current_datetime = datetime.now()
current_date = date.today()
```

```python
current_date.day
current_date.month
current_date.year
current_datetime.hour
current_datetime.minute 
current_datetime.second
current_datetime.microsecond
```

```python
today = datetime(2024,3,19,21,1,0)
today.hour
str(today + timedelta(days=14)) #'2024-04-02 21:01:00'
str(today - timedelta(days=14)) #'2024-03-05 21:01:00'

str(today - timedelta(days=14,hours=10)) #'2024-03-05 11:01:00'

str(datetime(2024,3,19,11,1,00) - datetime(2024,3,19,10,1,00))

delta = timedelta(days= 1, hours=3, minutes=30, seconds=30)
print(delta)

str(today - delta)
```

```python
datetime.strftime(today , '%Y-%m-%d %H:%M')
datetime.strptime('2024-03-19 21:07','%Y-%m-%d %H:%M')

datetime.strftime(today , '%d %B , %Y %H:%M %p')
```

```python
import requests
response = requests.get("https://api.thecatapi.com/v1/breeds")
response.text #returns the response contents in Unicode format.
response.content #returns the response contents in bytes.
response.status_code
response.headers.get("content-type")
```

```python
request = response.request 
request.url
request.path_url
request.method
request.headers
```

```python
headers = {"X-Request-Id":"12345"}
res = requests.get("https://example.org" , headers = headers)
res.request.headers
```

Requests contain relevant data regarding your API request call, such as the base URL, 
the endpoint, the method used, the headers, and so on. </br>
Responses contain relevant data returned by the server, including the data or content, 
the status code, and the headers.

200 - OK </br>
400 - Bad request  </br>
401 - UnAuthorized </br>
404 - Not Found  </br>
500 - Internal server error
